```python
import datetime
import sys
from google.cloud import storage

def get_presigned_url (bucket_name, blob_name) :
    # Create a client for interacting with the GCP Storage API, using the ServiceAccount key file
    client = storage.Client.from_service_account_json('/content/perqara-data-532572ce4996.json')
    # Create a bucket object
    bucket = client.bucket(bucket_name)

    blob = bucket.blob(blob_name)
    url = blob.generate_signed_url(
        version="v4",
        # Generate URL for 15 minutes
        expiration=datetime.timedelta(minutes=15),
        # Only Allow GET Call.
        method="GET",
    )
    print("Generated GET signed URL:")
    print(url)
```

```python
bucket_name = "perqara-ad-hoc"
blob_name = "consultation_chat_data/m_chat_messages_2024-01-17.csv"

get_presigned_url(bucket_name, blob_name)
```
