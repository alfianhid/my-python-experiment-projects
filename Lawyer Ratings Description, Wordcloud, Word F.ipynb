```python
from wordcloud import WordCloud

import matplotlib.pyplot as plt
```

```python
f = open("/content/desc_agg.txt", "r")

isi_text = f.read()
```

```python
wordcloud = WordCloud(width=1600, height=800, max_font_size=200, background_color='white')
```

```python
wordcloud.generate(isi_text)
```

```python
plt.figure(figsize=(12,10))

plt.imshow(wordcloud, interpolation='bilinear')

plt.axis("off")

plt.show()
```

```python
import re

frequency = {}
document_text = open('/content/desc_agg.txt')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)

for word in match_pattern:
    count = frequency.get(word, 0)
    frequency[word] = count + 1

sorted_frequency = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

for word, freq in sorted_frequency:
    print(word, freq)
```
