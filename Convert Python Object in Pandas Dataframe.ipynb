```python
import pandas as pd
import numpy as np
import ast
import json
import re
```

```python
file_path = "/content/notifications.json"

# Opening JSON file
f = open(file_path)
  
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
# Convert json data into pandas dataframe
df = pd.DataFrame(data)
# Convert 'summary' column to dictionary
df['data'] = df['data'].apply(lambda x: json.loads(x) if x else {})
# Flatten 'summary' column
summary_df = pd.json_normalize(df['data'])
# Prefix columns with 'summary_'
summary_df.columns = 'data_' + summary_df.columns
# Concatenate the summary_df DataFrame with the original df DataFrame
df = pd.concat([df.drop('data', axis=1), summary_df], axis=1)
```
