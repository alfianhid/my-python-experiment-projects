```python
import pandas as pd
import polars as pl
from google.cloud import storage
import io
import numpy as np
import math
from tqdm import tqdm
from pprint import pprint
import datetime
import numpy as np
```

```python
bucket_name = "perqara-dendrobium"

# Explicitly use service account credentials by specifying the private key file.
storage_client = storage.Client.from_service_account_json('/content/perqara-data-532572ce4996.json')

# Get the bucket and blob objects
bucket = storage_client.get_bucket(bucket_name)

# Download the contents of the blob as a string
source1 = "raw/mongodb/csv/m_chat_messages/m_chat_messages_2024-05-30_00.csv" # gs uri
blob1 = bucket.blob(source1)
str_data1 = io.BytesIO(blob1.download_as_string())

source2 = "raw/mongodb/csv/m_chat_messages/m_chat_messages_2024-05-30_06.csv" # gs uri
blob2 = bucket.blob(source2)
str_data2 = io.BytesIO(blob2.download_as_string())

source3 = "raw/mongodb/csv/m_chat_messages/m_chat_messages_2024-05-30_12.csv" # gs uri
blob3 = bucket.blob(source3)
str_data3 = io.BytesIO(blob3.download_as_string())
```

```python
source4 = "raw/postgres/csv/lawyers/lawyers.csv" # gs uri
blob4 = bucket.blob(source4)
str_data4 = io.BytesIO(blob4.download_as_string())

source5 = "raw/postgres/csv/lawyer_ratings/lawyer_ratings.csv" # gs uri
blob5 = bucket.blob(source5)
str_data5 = io.BytesIO(blob5.download_as_string())
```

```python
df = pl.read_csv(str_data1, separator='|')
df.head()
```

```python
df2 = pl.read_csv(str_data2, separator='|')
df2.head()
```

```python
df3 = pl.read_csv(str_data3, separator='|')
df3.head()
```

```python
df4 = pl.read_csv(str_data4, separator='|')
df4.head()
```

```python
df5 = pl.read_csv(str_data5, separator=',')
df5.head()
```

```python
filtered_lawyers_df = df4.select(["id", "slug", "agency_name"])
filtered_lawyers_df = filtered_lawyers_df.rename({"id": "lawyer_id", "slug": "lawyer_name"})
```

```python
merged_df = df5.join(filtered_lawyers_df, left_on="lawyer_id", right_on="lawyer_id", how="left")
merged_df.head()
```

```python
# merge data from df1, df2, and df3
merged_m_chat_messages = pl.concat([df, df2, df3], how="vertical_relaxed")
merged_m_chat_messages.head()
```

```python
# convert sent_at and read_at to Asia/Jakarta
columns_to_convert = ["sent_at", "read_at"]
for column in columns_to_convert:
    merged_m_chat_messages = merged_m_chat_messages.with_columns(
    pl.col(column)
    .str.strptime(pl.Datetime)
    .dt.replace_time_zone("UTC")
    .dt.convert_time_zone("Asia/Jakarta")
    .dt.strftime("%Y-%m-%d %H:%M:%S")
    .str.strptime(pl.Datetime, "%Y-%m-%d %H:%M:%S")
  )
```

```python
merged_m_chat_messages.head()
```

```python
lawyers_rating_below_4 = merged_df.filter((pl.col("rating") < 4) & (pl.col("rating") > 0))
lawyers_rating_below_4.head()
```

```python
# get filtered consultation_id from lawyer ratings
consultation_id_list = lawyers_rating_below_4.select("consultation_id").to_series().to_list()
```

```python
type(consultation_id_list)
```

```python
# filter chat data based on filtered consultation_id
filtered_chat_data = merged_m_chat_messages.filter(pl.col("consultation_id").is_in(consultation_id_list))
filtered_chat_data.head()
```

```python
# sort by consultation_id
filtered_chat_data = filtered_chat_data.sort("consultation_id")
filtered_chat_data.head()
```
