```python
import json
import pendulum
from pendulum import datetime
```

```python
key_list = ["message", "code"]
start = pendulum.datetime(2024, 8, 1)
end = pendulum.datetime(2024, 8, 10)
interval = pendulum.interval(start, end)
```

```python
with open("/content/response (2).json", "r") as dataFile:
  data = json.load(dataFile)

for key in key_list:
  if key in data:
    del data[key]

for ingestion_date in interval.range('days'):
  ingestion_date = ingestion_date.format("YYYY-MM-DD-HH-mm-ss")
  year, month, day, hour, minute, second = ingestion_date.split("-")
  my_datetime = datetime(int(year), int(month), int(day), int(hour), int(minute), int(second), 700000)
  utc_datetime = my_datetime.strftime("%Y-%m-%d %H:%M:%S.%f")
  data["data_datatime"] = utc_datetime

with open("/content/response_result_2.json", "w") as dataFile:
  data = json.dump(data, dataFile)
```
