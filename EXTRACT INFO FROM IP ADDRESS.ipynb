```python
import requests
import json
```

```python
def get_ip_geolocation(ip_address):
    url = f"http://ipinfo.io/{ip_address}/json?token=85ecf016ddeb40"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        return None
```

```python
def save_responses_to_ndjson(responses, output_file):
    # Write responses to a Newline Delimited JSON (NDJSON) file
    with open(output_file, 'w') as json_file:
        for response in responses:
            json.dump(response, json_file)
            json_file.write('\n')
```

```python
# List to store JSON responses
geolocation_responses = []

with open('/content/ip-jan-2024.txt') as ipfile:
    for single_ip in ipfile:
        ip_data = get_ip_geolocation(single_ip.strip())  # Remove newline characters
        if ip_data:
            geolocation_responses.append(ip_data)
```

```python
save_responses_to_ndjson(geolocation_responses, '/content/ip-jan-2024.json')
```
