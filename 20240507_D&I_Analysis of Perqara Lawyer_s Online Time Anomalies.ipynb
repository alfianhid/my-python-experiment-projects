```python
joined_all = pd.merge(joined_df, df_m_chat, how="left", on="consultation_id")
joined_all
```

```python
# Function to convert timedelta to string
def delta_to_string(delta):
    hours = delta.seconds // 3600
    minutes = (delta.seconds % 3600) // 60
    seconds = delta.seconds % 60
    return f"{hours} hours {minutes} minutes {seconds} seconds"
```

```python
# Convert timestamp to datetime object
joined_all.loc[:, 'sent_at'] = pd.to_datetime(joined_all['sent_at'])

# Add 7 hours to the timestamp column
joined_all.loc[:, 'sent_at'] = joined_all['sent_at'] + pd.Timedelta(hours=7)

# Calculate delta row by row
joined_all.loc[:, 'delta'] = joined_all['sent_at'].diff()

# Convert delta to seconds
joined_all['delta_seconds'] = joined_all['delta'].dt.total_seconds()
```

```python
# Convert delta to string
joined_all.loc[:, 'delta_str'] = joined_all['delta'].apply(delta_to_string)
result_df = joined_all[['consultation_id','user_name','sender_id','message','sent_at','delta','delta_seconds','delta_str']]
result_df.reset_index(drop=True, inplace=True)
result_df.dropna(subset=['sent_at'], inplace=True)
result_df.head()
```

```python
result_df = result_df[(result_df['delta_seconds'] > 0) & (result_df['delta_seconds'] <= 1800)]
```

```python
# Assuming df is your DataFrame
# Convert 'sent_at' column to datetime if it's not already in datetime format
result_df['sent_at'] = pd.to_datetime(result_df['sent_at'])

# Group by 'consultation_id'
grouped = result_df.groupby('consultation_id')
```

```python
# Function to calculate the time difference between first and last message
def calculate_time_difference(group):
    # Find indices of first and last message
    first_index = group.index[0]
    last_index = group.index[-1]

    # Retrieve 'sent_at' values corresponding to these indices
    first_sent_at = group.loc[first_index, 'sent_at']
    last_sent_at = group.loc[last_index, 'sent_at']

    # Calculate time difference
    time_difference = last_sent_at - first_sent_at
    return time_difference
```

```python
# Apply the function to each group and get the result as a DataFrame
time_differences = grouped.apply(calculate_time_difference)

# Display the result
print(time_differences)
```

```python
# Convert Series to DataFrame
time_differences_df = time_differences.reset_index()

# Rename the columns for clarity
time_differences_df.columns = ['consultation_id', 'time_difference']

# Display the resulting DataFrame
time_differences_df
```
