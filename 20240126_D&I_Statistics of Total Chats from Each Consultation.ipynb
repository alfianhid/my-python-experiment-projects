```python
import polars as pl

source = "/content/raw_mongodb_pq_m_chat_messages.parquet"

df = pl.read_parquet(source)
```

```python
df.head()
```

```python
grouped_df = df.group_by("consultation_id", maintain_order=True).agg(pl.col("message"))
```

```python
grouped_df.head()
```

```python
type(grouped_df)
```

```python
custom_df = grouped_df.select(
    "*",  # Select all existing columns
    pl.col("message").map_elements(lambda messages: len(messages)).alias("message_count")
)
```

```python
custom_df
```

```python
# Define a function to calculate word and character count
def calculate_word_counts(message_list):
    word_count = 0
    for message_text in message_list:
        word_count += len(message_text.split())

    return word_count

def calculate_char_counts(message_list):
    char_count = 0
    for message_text in message_list:
        char_count += len(message_text)

    return char_count

# Apply the function to create new columns
special_df = custom_df.select([
    pl.col("consultation_id"),
    pl.col("message"),
    pl.col("message_count"),
    pl.col("message").map_elements(calculate_word_counts, return_dtype=pl.UInt32).alias("word_count"),
    pl.col("message").map_elements(calculate_char_counts, return_dtype=pl.UInt32).alias("char_count")
])
```

```python
special_df
```

```python
avg_word_count = special_df.select(pl.col("word_count").mean())
print(avg_word_count)
```

```python
med_word_count = special_df.select(pl.col("word_count").median())
print(med_word_count)

quan1_word_count = special_df.select(pl.col("word_count").quantile(0.3))
print(quan1_word_count)

quan3_word_count = special_df.select(pl.col("word_count").quantile(0.7))
print(quan3_word_count)
```

```python
avg_char_count = special_df.select(pl.col("char_count").mean())
print(avg_char_count)
```

```python
med_char_count = special_df.select(pl.col("char_count").median())
print(med_char_count)

quan1_char_count = special_df.select(pl.col("char_count").quantile(0.3))
print(quan1_char_count)

quan3_char_count = special_df.select(pl.col("char_count").quantile(0.7))
print(quan3_char_count)
```

```python
avg_message_count = custom_df.select(pl.col("message_count").mean())
print(avg_message_count)
```

```python
med_message_count = custom_df.select(pl.col("message_count").median())
print(med_message_count)
```

```python
quan1_message_count = custom_df.select(pl.col("message_count").quantile(0.3))
print(quan1_message_count)
```

```python
from datetime import datetime
```

```python
not_null_sent_at_df = df.filter(
    ~(pl.col("sent_at") == datetime(1970, 1, 1))
)
print("NOT NULL sent_at count is", not_null_sent_at_df.shape[0])
print("Percentage:", ((not_null_sent_at_df.shape[0] * 100)/df.shape[0]))
```

```python
null_sent_at_df = df.filter(
    (pl.col("sent_at") == datetime(1970, 1, 1))
)
print("NULL sent_at count is", null_sent_at_df.shape[0])
print("Percentage:", ((null_sent_at_df.shape[0] * 100)/df.shape[0]))
```
